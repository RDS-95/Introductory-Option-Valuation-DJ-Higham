#### Interest Rates

continuously compounded interest rate formula:

D(t) = e^(rt)D(0)

let r denote the annual rate here.

This implies that the following offers are of equal value

a) 100 immediately

b) 100e(rt) at time t

A deal which is worth 100 at time t will then be worth 100e(-rt) at time 0. This is discounting for inflation/interest

#### Short Selling

Portfolio: Combination of assets, options, cash

Owning a -ve amount of cash => cash has been borrowed

Owning a -ve amount of assets/options is possible through short selling

Short selling: selling an item that is not owned with the intention of buying it back at a later date

to demo: S(t) value of asset at time t

For short selling, gain an amount S(t1) from the short sale, pay an amount S(t2) from the buy back

overall profit loss

e^(r(t2-t1) S(t1)-S(t2)

#### Arbitrage

There is never an opportunity to make a risk free profit that gives a greater return that that provided by the interest from a bank deposit.

Principle of no arbitrage.

arbitrageurs scour markets to find better than interest rate level opportunities.

#### Put-Call Parity

Defines a relationship between value C of euro call option and value P of euro put option. strike price E, expiry date T

two portfolios:

PiA: call option + Ee^(rt)

PiB: put option + unit of asset

@ expiry date, both are worth the same. At time 0, both are worth the same, so:

C + Ee^(-rT) = P + S

so if we can value a call, we can value a put.

this would only not be the case if we are violating the no arbitrage principle

#### Upper and Lower Bounds on Option Values

We can use similar arguments to obtain bounds on C and P. Study the call option.

PiA: one call option plus Ee^(-rT) cash

PiB: one unit of asset

PiA has payoff max(S(T),E) on expiry. PiB has payoff S(T). So PiB must have a zero time value no greater than PiA.

C >= S - Ee^(-rT)

call option can't have a negative value so we can write

C >= max((S - Ee^(-rT)),0)

since strike price is always >=0, 

C =< S

So the upper, lower bounds for P:

P >= max(Ee^(-rT)-S,0)  P =< Ee^(-rT)

The time-zero European Call option value, C is nondecreasing as a function of the expiry date T. useful proof in text.

#### What are Options?

Asset: any financial object whose value is known but is liable to change

DEF
A European call option gives its holder the right (but not the obligation) to purchase from the writer a prescribed asset
for a prescribed price at a prescribed time in the future.

prescribed purchase price: Exercise/Strike price
future time: expiry date

the opposite of a call option is a put option

DEF
A European put option gives its holder the right to sell to the writer a prescribed asset for a prescribed price
at a prescribed time in the future.

How much should the holder pay for the priviledge? We need to derive a model for the asset price to determine this

#### Why do we study Options?

options are extremely attractive to investors, for speculation and hedging
there is a systematic method to determine their value so they can be traded with certainty

By combining different types of option, an investor can take a position that reaps benefits from various types of asset behaviour

We can visualize options with payoff diagrams

Value of european call option:
C = max(S(T)-E,0) 

Value of european put option:
P = max(E-S(T),0)

#### How are options traded?

Options can be traded on official exchanges. CBOE was the first in 1973

Exchanges operate through the use of market makers. individuals who are obliged to buy or sell options whenever asked.

market maker on request will quote a price for the option. Or, more precisely, two prices. The bid and the ask.

the bid: price at which market maker will buy the option from you
the ask: price at which market maker will sell the option to you.

bid < ask. difference is the bid-ask spread.

OTC deals are trades between financial institutions.

Details of option prices are often published in the financial times

#### Typical Option Prices

American call equal in value to european
American put higher value to european

Increasing the strike price has a negative effect on the payoff and therefore decreases a call options worth.

For a given strike price, both the call and the put option prices increase when time to expiry increases.

#### Other Financial Derivatives

European call and put options are classic examples of financial derivatives.

Derivative means that their value is derived from an underlying asset.

Develop our understanding using these european options before moving on to other exotic options

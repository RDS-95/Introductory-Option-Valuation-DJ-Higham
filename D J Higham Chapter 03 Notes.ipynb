#### Random Variables, probability and mean

discrete random variable: takes values from a finite set of numbers

set of probabilities is associated with each random variable

Bernoulli random variable => variable X taking the value 1 with probability p and 0 with probability 1-p

continuous random variable can take any value in reals

X continuous => density function such that the probability of a =< X =< b found by integrating f(x) from a to b

uniform distribution from a-b means the density function takes the same value between a and b. and it is equally liekly to take any of these values => X~U(a,b)

for random variables, E(X)+E(Y)=E(X+Y), E(aX)=aE(X) for a in R



#### Independence

#### Normal Distribution

standard normal distribution => X is a continuous random variable with the density function: 

f(x) = (1/(sqrt(2*pi)))(e^((-x^2)/2))

the distribution function then is the area under the density curve to the left of x

N(x) := distribution function for standard normal random variable.

#### Central Limit Theorem

The central limit theorem states that if you have a population with mean μ and standard deviation σ and take sufficiently large random samples from the population with replacement , then the distribution of the sample means will be approximately normally distributed.

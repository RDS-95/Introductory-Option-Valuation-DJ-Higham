```python
#an illustration of compound interest

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

Dzero = 5
r = 0.15
T = 5
m = 12

Di = Dzero*(pow((1+r/m), m*T))
```

```python
#COMPOUNDED MONTHLY

def CompoundIntCalc():
    Dzero = int(input("Enter starting principal: "))
    m = 12
    r = float(input("Enter annual interest rate: "))
    T = int(input("Enter the amount of years: "))
    Di = Dzero*(pow((1+r/m), m*T))
    print('Value at end of term: ', Di)
    
    results = pd.DataFrame(columns = ['Year', 'Amount'])
    for i in range(1,T+1):
            Year = i
            Amount = Dzero*np.power((1 + r / m), m * i)
            results =  results.append({'Year': Year, 'Amount': Amount}, ignore_index = True)
            results['Year'] = results['Year'].astype('int')
        
    plt.plot(results['Year'],results['Amount'])
    plt.grid()
```

```python
CompoundIntCalc()
```

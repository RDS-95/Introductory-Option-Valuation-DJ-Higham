#### Pseudo Random Numbers

It is normally sufficient to work with pseudo random numbers for this book.

In python we can generate normally distributed and unifromly distributed random numbers as so:

```python
import random
import numpy as np
import matplotlib.pyplot as plt
```

```python
gaussianlist = []

for i in range(0,9):
    n = random.gauss(0,1)
    gaussianlist.append(n)

print(gaussianlist)
```

```python
uniformlist = []

for i in range(0,9):
    n = random.random()
    uniformlist.append(n)
    
print(uniformlist)
```

#### Statistical Tests

We can test a pseudo random number generator by taking M samples and computing the sample mean (arithmetic average of sample values) and the sample variance (expected value defining it's variance).

```python
#m number of samples on a uniform distribution from 0-1

def mean_sample_calc(m):
    uniformlist = []
    for i in range(0,m):
        i = random.random()
        uniformlist.append(i)
    print('Mean of', m, 'sample distribution: {}'.format(np.mean(uniformlist)))   
```

```python
mean_sample_calc(100)
```

```python
#Another function to plot mean as m goes up?

def plot_mean_sample_calc(p):
    meanlist = []
    for i in range (1,p):
        mean_sample_calc(i)
        x = np.mean(uniformlist)
        meanlist.append(x)
    plt.xlabel('sample size')
    plt.ylabel('mean')
    plt.ylim(0.45,0.55)
    plt.plot(range(1,p), meanlist)
```

```python
plot_mean_sample_calc(10)
```

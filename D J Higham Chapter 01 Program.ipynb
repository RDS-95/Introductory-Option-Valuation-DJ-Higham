```python
#We are building a program to make a bull spread payoff diagram

import numpy as np
import matplotlib.pyplot as plt

E1 = 2
E2 = 4

S = np.linspace(0,6,100)
B = np.maximum(S-E1,0) - np.maximum(S-E2,0)

plt.plot(S,B)
plt.title('Bull Spread Payoff')
plt.axis([0,6,-0.1,2.1])
```

```python
def BullSpreadPayoff(E1, E2):
    S = np.linspace(0,6,100)
    B = np.maximum(S-E1,0) - np.maximum(S-E2,0)
    plt.plot(S,B)
    plt.title('Bull Spread Payoff')
```

```python
BullSpreadPayoff(2,4)
```

```python
#Now a butterfly spread payoff

def ButterflySpreadPayoff(E1, E3):
    S = np.linspace(0,6,100)
    E2 = (E1 + E3)/2
    B = np.maximum(S-E1,0) + np.maximum(S-E3,0) - 2*(np.maximum(S-E2,0))
    plt.plot(S,B)
    plt.title('Butterfly Spread Payoff')
```

```python
ButterflySpreadPayoff(2,4)
```
